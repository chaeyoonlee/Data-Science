### Joining
** Equivalent SQL query:**  
Select *  
From CityTemps, TempsRegions  
Where CityTemps.state = TempsRegions.state

```python
TR_dataframe = pd.read_csv("TempsRegions.csv")
join = CT_dataframe.merge(TR_dataframe, on='state')
```

### Group By
