# Modify iris for clustering such that target is removed

```r
newiris <- iris
newiris$Species <- NULL
```

# Apply K-means with 3 clusters and print statistics

```r
(kc <- kmeans(newiris, 3))
```

# Compare species label with clustering results
# 1:virginica , 2:versicolor,  3:setosa

```r
table(iris$Species, kc$cluster)
```

```r
kc$centers
kc$cluster
```

```r
# Plot the points on the map, use col=kc$cluster to apply a color based on the cluster group
plot(newiris[c("Sepal.Length", "Sepal.Width")], col=kc$cluster)
# (optional) plot the centers for each cluster 3개로 clustrering했으니까 col=1:3
# col:color(different color for each center), pch:shape 8이라 별모양 저거 있음, cex:size
points(kc$centers[,c("Sepal.Length", "Sepal.Width")], col=1:3, pch=8, cex=2)
```
